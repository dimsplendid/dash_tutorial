```python
import plotly.express as px
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output
from dash.exceptions import PreventUpdate
import pandas as pd
```

```python
def load_data():
    return pd.read_excel('./test_data/test_data_RD11001105.xlsx')

df = load_data() # become database query in the future
df.head()
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    html.Div([
        "LC: ",
        lc_selection := dcc.Dropdown(options=df.LC.unique(), multi=True),
        "Chips: ",
        chip_selection := dcc.Dropdown(multi=True),
        "VT Curve",
        vt_fig := dcc.Graph(),
    ]),
    result := html.Div(),
])

@app.callback(
    Output(chip_selection, component_property='options'),
    Input(lc_selection, component_property='value'),   
)
def update_chip_options(selected_lcs):
    if selected_lcs is None:
        # PreventUpdate prevents ALL outputs updating
        raise PreventUpdate
    selected_df = df[df['LC'].isin(selected_lcs)]
    return [{'label': x, 'value': x} for x in selected_df.ID.unique()]


@app.callback(
    Output(vt_fig, component_property='figure'),
    Input(chip_selection, component_property='value'),
)
def update_vt_curve(selected_chips):
    if selected_chips is None:
    # PreventUpdate prevents ALL outputs updating
        raise PreventUpdate
    selected_df = df[df['ID'].isin(selected_chips)]
    fig = px.scatter(selected_df, x='Vop', y='LC%', color='ID')
    fig.update_layout(transition_duration=200)
    return fig

@app.callback(
    Output(result, component_property='children'),
    Input(lc_selection, component_property='value'),
)
def update_output_div(input_value):
    return f'LC Selection: {input_value}, type: {type(input_value)}'

app.run_server(mode='inline')
```
